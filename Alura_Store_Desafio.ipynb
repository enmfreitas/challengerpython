#1. Análise do faturamento



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
faturamento_lojas = {
    "Loja 1": loja1["Preço"].sum(),
    "Loja 2": loja2["Preço"].sum(),
    "Loja 3": loja3["Preço"].sum(),
    "Loja 4": loja4["Preço"].sum()
}

faturamento_lojas
```

# 2. Vendas por Categoria

```python
# Ver categorias mais vendidas em cada loja
for i, loja in enumerate([loja1, loja2, loja3, loja4], start=1):
    print(f"\n=== Loja {i} ===")
    print(loja["Categoria do Produto"].value_counts())
```

# 3. Média de Avaliação das Lojas


```python
# Média das avaliações de cada loja
media_avaliacao = {
    "Loja 1": loja1["Avaliação da compra"].mean(),
    "Loja 2": loja2["Avaliação da compra"].mean(),
    "Loja 3": loja3["Avaliação da compra"].mean(),
    "Loja 4": loja4["Avaliação da compra"].mean()
}

media_avaliacao
```

# 4. Produtos Mais e Menos Vendidos

```python
# Produtos mais e menos vendidos em cada loja
for i, loja in enumerate([loja1, loja2, loja3, loja4], start=1):
    print(f"\n=== Loja {i} ===")
    produtos_vendidos = loja["Produto"].value_counts()
    produto_mais = produtos_vendidos.idxmax()
    qtd_mais = produtos_vendidos.max()
    produto_menos = produtos_vendidos.idxmin()
    qtd_menos = produtos_vendidos.min()
    print(f"Mais vendido: {produto_mais} ({qtd_mais} vendas)")
    print(f"Menos vendido: {produto_menos} ({qtd_menos} vendas)")
```

# Relatório Final – Desafio Alura Store

## Objetivo
O objetivo deste projeto foi identificar, com base em dados reais de vendas, qual das quatro lojas da **Alura Store** apresenta o menor desempenho e, portanto, é a mais indicada para ser vendida pelo Sr. João.

## Metodologia
As análises foram realizadas utilizando Python e as bibliotecas **Pandas** e **Matplotlib** no Google Colab.  
Foram consideradas as seguintes métricas:

- Faturamento total por loja;  
- Categorias mais vendidas;  
- Média de avaliação dos clientes;  
- Produtos mais e menos vendidos;  
- Frete médio por loja.

## 1. Faturamento Total
O gráfico mostrou que a **Loja 3** apresentou o maior faturamento, enquanto a **Loja 4** teve o menor.  
Essa diferença indica que a Loja 4 tem menor volume de vendas e menor contribuição para a receita geral.

## 2. Categorias Mais Vendidas
As categorias mais populares variaram entre as lojas, mas observou-se destaque para **eletrônicos** e **móveis** em volume de vendas.  
Essas categorias representam os maiores volumes de faturamento em praticamente todas as lojas.

## 3. Média de Avaliação dos Clientes
As médias de avaliação variaram entre **3,8 e 4,6**.  
A **Loja 3** obteve a melhor média (4,6), enquanto a **Loja 2** apresentou a pior (3,8), sugerindo maior insatisfação dos clientes.

## 4. Produtos Mais e Menos Vendidos
- Produtos mais vendidos: *Smartwatch*, *Jogo de tabuleiro*, *Kit banquetas* e *Cama box*.  
- Produtos menos vendidos: *Impressora 3D* e *Guitarra*, com volumes muito baixos de venda.  
Esses dados ajudam a compreender a composição das vendas e possíveis produtos de baixa rotação.

## 5. Frete Médio
O custo médio de frete variou entre **29 e 34 reais**, com a **Loja 3** apresentando o valor mais alto (≈34 R$).  
Apesar disso, a Loja 3 mantém o maior faturamento e a melhor média de avaliações — indicando boa aceitação mesmo com custo de entrega superior.

## Conclusão
Após a análise dos dados e métricas:

- **Loja 3** é a mais lucrativa e bem avaliada;  
- **Loja 1 e Loja 2** têm desempenho intermediário;  
- **Loja 4** apresenta **o pior desempenho geral**, com menor faturamento e vendas menos expressivas.

**Recomendação:**  
O Sr. João deve considerar **vender a Loja 4**, pois ela é a que menos contribui para o faturamento total e demonstra desempenho inferior em múltiplos indicadores.
